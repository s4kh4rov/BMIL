```python
def div_cols(df):
    df_division = {'number': [], 'string': []}
    for i in df.columns:
        # Numeric data
        if df[i].dtype=='int64':
            df_division['number'].append(i)
        elif df[i].dtype=='float64':
            df_division['number'].append(i)
        # Categorical data
        else:
            df_division['string'].append(i)
    return df_division
div_cols(data)
```
